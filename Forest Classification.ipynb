# Dataset 1: Iris

```python
import pandas as pd
```

```python
#Import data and analyse:
data= pd.read_csv('Iris.csv') #Iris dataset gives us the 3 species of iris flower based upon the size of petals. It CATEGORIZES the Flower in 3 Species.
data.head()
```

```python
data.info()
```

```python
data['Species'].value_counts()
```

```python
#Choosing the model:
X=data.iloc[:,[1,2,3,4]].values
Y=data.iloc[:,5].values
X
```

```python
#Transforming the Y using Label Binarizer:
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()

Y=lb.fit_transform(Y) #gives Y as 0's and 1's representing each species by different arrangements of 1
Y
```

```python
#Train and Test variables:
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.2)
```

```python
#Using the Forest Classification by Random Forest Classifier:
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()

rf.fit(X_train,Y_train)
```

```python
#Checking the accuracy:
accuracy=rf.score(X_test,Y_test)
accuracy
```

```python
#Getting new data and predicting the species(classification):
new_data=[5.1, 3.4, 1.5, 0.2]

predict=rf.predict([new_data])

if (predict[0]==[1,0,0]).all:
    print('Iris-setosa')
elif (predict[0]==[0,1,0]).all:
    print('Iris-versicolor')
elif (predict[0]==[0,0,1]).all:
    print('Iris-virginica')
else:
    print('Invalid')
```

# Dataset 2: Loan Approval

```python
my_data=pd.read_csv('loan_approval_dataset.csv')
my_data.head()
```

```python
my_data.info()
```

```python
x=my_data.iloc[:,[1,4,5,6,7,8,9,10,11]].values
y=my_data.iloc[:,12].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
y=lb.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier()
rc.fit(x_train,y_train)
```

```python
rc.score(x_test,y_test)
```

```python
result= rc.predict([[0,4100000,12200000,8,417,2700000,2200000,8800000,3300000]])

if (result[0]==0):
    print('Loan would be Approved')
else:
    print('Loan would not be approved')
```
