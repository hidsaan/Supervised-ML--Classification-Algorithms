# Dataset 1: Diabetes

```python
import pandas as pd
```

```python
data=pd.read_csv('diabetes.csv')
data.head()
```

```python
data.info()
```

```python
X=data.drop(['Outcome'],  axis=1)
Y=data['Outcome']
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
#Checking the accuracy using confusion matrix: Step 1- Make a predict variable:
Y_pred= dtc.predict(X_test) #We compare the Y_test values with Y_pred using X_test as inpuy (Since X_test provides inputs for Y_test as well)
```

```python
#Step 2: Actually using the confusion matrix:
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,Y_pred) #Comparison drawn between Y_test and Y_pred in a comparison matrix
#the 66 is the Y_test value for true positives, and 21 is the true negatives. 42 is the false positives from Y_pred, and 25 the false negatives
```

```python
new_array=[6,148,72,35,0,33.6,0.627,50]
predict=dtc.predict([new_array])

if (predict[0]==0):
    print("The person is non-diabetic")
else:
    print("The person is diabetic")
```

# Dataset 2: Iris

```python
my_data=pd.read_csv('Iris.csv')
my_data.head()
```

```python
x=my_data.iloc[:,[1,2,3,4]].values
y=my_data.iloc[:,5].values
x
```

```python
from sklearn.preprocessing import LabelBinarizer
lb= LabelBinarizer()
y=lb.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
classifier.score(x_test,y_test)
```

```python
result=classifier.predict([[5.9,4.2,2.6,0.9]])

if (result[0]==[1,0,0]).all:
    print('Iris-setosa')
elif (result[0]==[0,1,0]).all:
    print('Iris-versicolor')
elif (result[0]==[0,0,1]).all:
    print('Iris-virginica')
else:
    print('invalid')
```
