# Dataset 1: Social Network Marketing

```python
import pandas as pd
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
X=data.iloc[:,[1,2,3]].values
Y=data.iloc[:,4].values
X
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
X[:,0]=lb.fit_transform(X[:,0])
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.3, random_state=0) #random state defines how algorithm can take any values for training and testing upon rerendering. 0 turns it off
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(X_train,Y_train)
```

```python
classifier.score(X_test,Y_test)
```

```python
Y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_pred,Y_test)
```

```python
new_data=['Female',35,70000]
new_data[0]=lb.fit_transform([new_data[0]])[0]

result=classifier.predict([new_data])

if (result[0]==1):
    print('The person will buy the product')
else:
    print('The person will not buy the product')
```

# Dataset 2: Haberman's Experiment of Life Expectancy of Cancer Patients

```python
my_data=pd.read_csv('haberman1.csv')
my_data.head()
```

```python
my_data.info()
```

```python
x=my_data.iloc[:,[0,1,2]].values
y=my_data.iloc[:,3].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
```

```python
gb.score(x_test,y_test)
```

```python
result= gb.predict([[20,22,5]])

if(result[0]==1):
    print('Survival chances of 5 years or longer')
else:
    print('Survival chances of less than 5 years')
```
